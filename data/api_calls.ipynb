```python
import pandas as pd
import requests
```

```python
df = pd.read_csv("aefis_courses.csv")
courses_list = []
df.head()
```

```python
for index, row in df.iterrows():
    courses_list.append(row['Subject Code'] + " " + str(row['Course Number']))
```

```python
len(courses_list)
```

```python
course_grades_dict = {}
```

```python
for course in courses_list:
    try:
        r = requests.get('https://api.madgrades.com/v1/courses', headers={'Authorization': 'Token token=<TOKEN>'}, params={'query':course}).json()
        r = requests.get(r['results'][0]['url']+'/grades', headers={'Authorization': 'Token token=<TOKEN>'}).json()
        r_cum = r['cumulative']
        r_sum = r_cum['aCount']+r_cum['abCount']+r_cum['bCount']+r_cum['bcCount']+r_cum['cCount']+r_cum['dCount']+r_cum['fCount']
        if(r_sum != 0):
            r_gpa = ((r_cum['aCount']*4.0)+(r_cum['abCount']*3.5)+(r_cum['bCount']*3.0)+(r_cum['bcCount']*2.5)+(r_cum['cCount']*2.0)+(r_cum['dCount']*1.0)+(r_cum['fCount']*0))/(r_sum)
        else:
            r_gpa = 0
    except Exception as e:
        r_gpa = 0
        
    course_grades_dict[course] = r_gpa
    print(course +": " + str(r_gpa))
```

```python
import json

with open('course_grades.txt', 'w') as convert_file:
     convert_file.write(json.dumps(course_grades_dict))
```

```python
for course in course_grades_dict:
    course_grades_dict[course] = round(course_grades_dict[course], 2)
```

```python
#final
import json

with open('CourseGrades.txt', 'w') as convert_file:
     convert_file.write(json.dumps(course_grades_dict))
```
